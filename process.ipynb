```python
import pandas as pd
import numpy as np

np.random.seed(42)
```

```python
df = pd.read_csv('data.csv')
```

```python
df_nans = df.copy()

for col in df_nans.columns:
    if col == 'popularity': continue

    mask = np.random.rand(len(df_nans)) < 0.07
    df_nans.loc[mask, col] = np.nan

df_nans.head(10)
```

```python
df_nans.to_csv('data_nans.csv')
```

```python
group_ids = np.random.randint(0, 3, size=len(df_nans))

for i in range(3):
    part = df_nans[group_ids == i]
    part.to_csv(f"data/dataset_{i+1}.csv", index=False)
```
